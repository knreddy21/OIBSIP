```python
df.drop_duplicates(inplace=True)
```

```python
df['name'].fillna(df['name'].mode()[0], inplace=True)
df['host_name'].fillna(df['host_name'].mode()[0], inplace=True)
```

```python
df.isnull().sum()
```

```python
df['reviews_per_month'].fillna(0, inplace=True)
```

```python
df.isnull().sum()
```

```python
# Ensure datetime
df['last_review'] = pd.to_datetime(df['last_review'], errors='coerce')

# Find min & max (already known, but keeping dynamic)
min_date = df['last_review'].min()
max_date = df['last_review'].max()

# Count missing values
missing_count = df['last_review'].isnull().sum()

# Generate random dates between min & max
random_dates = pd.to_datetime(
    np.random.randint(
        min_date.value // 10**9,
        max_date.value // 10**9,
        missing_count
    ),
    unit='s'
)

# Fill missing values
df.loc[df['last_review'].isnull(), 'last_review'] = random_dates
```

```python
df.isnull().sum()
```

```python
# Convert text to lowercase & remove extra spaces
df['name'] = df['name'].str.lower().str.strip()
df['host_name'] = df['host_name'].str.lower().str.strip()
df['room_type'] = df['room_type'].str.lower().str.strip()
df['neighbourhood_group'] = df['neighbourhood_group'].str.lower().str.strip()
```

```python
# Ensure date format is uniform
df['last_review'] = pd.to_datetime(df['last_review'], errors='coerce')
```

```python
# Min-Max scaling for price
df['price_scaled'] = (df['price'] - df['price'].min()) / \
                     (df['price'].max() - df['price'].min())
```

```python
Q1 = df['price'].quantile(0.25)
Q3 = df['price'].quantile(0.75)
IQR = Q3 - Q1
```

```python
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

```python
df = df[(df['price'] >= lower_bound) & (df['price'] <= upper_bound)]
```

```python
df['price'].describe()
```

```python
import matplotlib.pyplot as plt

plt.boxplot(df['price'])
plt.title("Price Distribution After Outlier Removal")
plt.ylabel("Price")
plt.show()
```
