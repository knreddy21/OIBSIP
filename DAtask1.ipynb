```python
import pandas as pd

# Step 1: Read data
df = pd.read_csv("C:\\Users\\spred\\OneDrive\\Documents\\retail_sales_dataset.csv")   # your file name

# Step 2: Convert Date column correctly
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

# Step 3: Create Month column (SAFE WAY)
df['Month'] = df['Date'].dt.to_period('M')

# Step 4: Monthly sales per product
monthly_product_sales = df.groupby(
    ['Month', 'Product Category']
)['Total Amount'].sum().reset_index()
monthly_product_sales['Month']=monthly_product_sales['Month'].astype(str)

# Step 5: Show result
print(monthly_product_sales)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Convert Date column to datetime
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

# 2. Create Month column
df['Month'] = df['Date'].dt.month_name()

# 3. Create Month order (important for sorting)
month_order = ['January','February','March','April','May','June',
               'July','August','September','October','November','December']

df['Month'] = pd.Categorical(df['Month'], categories=month_order, ordered=True)

# 4. Group data: Product vs Month
monthly_product_sales = df.groupby(['Month', 'Product Category'])['Total Amount'].sum().reset_index()

# 5. Pivot table (best for visualization)
pivot_table = monthly_product_sales.pivot(index='Month',
                                           columns='Product Category',
                                           values='Total Amount')

# 6. Visualization
pivot_table.plot(kind='bar', figsize=(14,6))

plt.title("Monthly Sales of Each Product")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.legend(title="Product Category")
plt.tight_layout()
plt.show()
```
